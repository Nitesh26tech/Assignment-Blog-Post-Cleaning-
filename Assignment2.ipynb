```python
with open("messy_blog.txt", 'r') as R: 
    messyblog = R.read()
    print(messyblog)
```

```python
# print 100 characters
# Char = messyblog[:100]
reversChar = messyblog[-1:100]
print(reversChar) 
# print(len(Char))
```

```python
# remove extra spaces
# print(blog.strip())
```

```python
#remove unwanted symbols
cleanText = messyblog.replace("#","").replace("@","").replace("!","").replace("?","").replace("...","")
# print(cleanText)
replaceWord = cleanText.replace("gr8","great").replace("btw","by the way").replace("idk","i dont know").replace("lol","loughing out loud")
# print(replaceWord)
lowercase = replaceWord.lower()
print(lowercase)
```

```python
# fix spacing between words
# print("".join(lowercase))
# print(lowercase.split())

replaceLine = " ".join(lowercase.split())
# print(replaceLine)

capital = replaceLine.capitalize()
print(capital)
```

```python
with open ("clean_blog.txt", "w") as f:
    f.write("")
with open ("clean_blog.txt", "a") as f:
    f.write(capital)
    print("Blog is clean successfully!! \n")
    print(" Blog is saved as clean_blog.txt ")
```
